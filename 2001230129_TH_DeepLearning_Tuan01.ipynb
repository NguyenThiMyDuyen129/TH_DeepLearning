**Bài 3: Tìm hiểu hàm numpy.greater, numpy.greater_equal, numpy.less, numpy.less_equal để so sánh 2 mảng 1 chiều, 2 chiều**
- numpy.greater (x1, x2) x1 > x2: Kiểm tra phần tử của mảng 1 có lớn hơn mảng 2 không.
- numpy.greater_equal(x1,x2) x1 >= x2: Kiểm tra phần tử của mảng 1 có lớn hơn hoặc bằng mảng 2 không.
- np.less(x1, x2)	x1 < x2: Kiểm tra phần tử của mảng 1 có nhỏ hơn mảng 2 không.
- np.less_equal(x1, x2)	x1 <= x2	Kiểm tra phần tử của mảng 1 có nhỏ hơn hoặc bằng mảng 2 không.

```python
import numpy as np

a = np.array([1, 5, 10])
b = np.array([2, 5, 8])

print("Greater (a > b):      ", np.greater(a, b))
print("Greater Equal (a >= b):", np.greater_equal(a, b))
print("Less (a < b):         ", np.less(a, b))
print("Less Equal (a <= b):   ", np.less_equal(a, b))
```

```python
a = np.array([[1, 10], [20, 30]])
b = np.array([[5, 5], [20, 25]])

# So sánh m1 >= m2
sosanh = np.greater_equal(a, b)

print("Ma trận 1:\n", a)
print("Ma trận 2:\n", b)
print("Kết quả (m1 >= m2):\n", sosanh)
```

**Bài 12: Viết chương trình tạo mảng 3x3x3 và tính tổng theo mỗi dòng, mỗi cột**

```python
array = np.random.randint(1, 10, size=(3, 3, 3))
print('Ma trận random (3x3x3)\n', array)
print('-' * 30)
# 2. Tính tổng theo mỗi dòng (axis=2)
# Mỗi dòng nằm trên phương ngang của từng 'lớp'
sum_rows = np.sum(array, axis=2)

# 3. Tính tổng theo mỗi cột (axis=1)
# Mỗi cột nằm trên phương dọc của từng 'lớp'
sum_cols = np.sum(array, axis=1)

print("Tổng theo mỗi dòng (kết quả là ma trận 3x3):\n", sum_rows)
print("\nTổng theo mỗi cột (kết quả là ma trận 3x3):\n", sum_cols)
```

**Bài 13: Tạo 2 vector ngẫu nhiên từ 10 phần tử. Thực hiện inner product(np.dot)**

```python
#Tạo random 2 vector 10 phần tử [0,10]
v1 = np.random.randint(0, 11, 10)
v2 = np.random.randint(0, 11, 10)

print("Vector 1:", v1)
print("Vector 2:", v2)
print("-" * 30)

#Thực hiện Inner Product bằng np.dot()
inner_product = np.dot(v1, v2)

#inner_product = v1 @ v2
print("Kết quả Inner Product (Tích vô hướng):", inner_product)
```
